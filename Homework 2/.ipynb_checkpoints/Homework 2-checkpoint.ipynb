1. распределение qsec в виде гладкой линии и гистограммы (hist, plot-density-polygon)

```r
df <- mtcars
head(df)
```

```r
hist(mtcars$qsec, 
     breaks = 10,
     col="skyblue", 
     border="black", 
     xlab="qsec", 
     main="Histogram of qsec", 
     sub="Distribution of qsec", 
     cex.lab=1.5, 
     cex.axis=1.5, 
     cex.main = 2
    )
```

```r
plot(density(mtcars$qsec), 
     col = "red", 
     main = "Плотность распределения qsec", 
     xlab = "qsec", 
     ylab = "Плотность", 
     cex.lab = 2, 
     cex.axis = 1.5, 
     cex.main = 2
    )

polygon(density(mtcars$qsec), col = rgb(1, 0, 0, 0.2))
```

2. Точечный график, на котором будет отображаться зависимость mpg от disp (plot). Задать свой цвет, форму элементов для каждого значения cyl. Размер элементов должен быть равен 2. 

```r
plot(mtcars$disp, 
     mtcars$mpg, 
     col=mtcars$cyl, 
     main = 'Зависимость mpg от disp', 
     pch=c(1,3,5),
     cex=2, 
     xlab="disp", 
     ylab="mpg", 
     cex.lab = 2, 
     cex.axis = 1.5, 
     cex.main = 2
    )

legend("topright", 
       legend=unique(mtcars$cyl),
       col=unique(mtcars$cyl), 
       pch=c(1,3,5),
       title="cyl", 
       cex = 1.75
      )
```
