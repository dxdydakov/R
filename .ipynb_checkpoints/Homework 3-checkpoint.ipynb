Аналогично заданию №2. Визуализировать с помощью ggplot2.

Для датасета mtcars визуализировать:

1. распределение qsec в виде гладкой линии и гистограммы (hist, plot-density-polygon)

```r
df <- mtcars
```

```r
library(ggplot2)
```

```r
ggplot(df, aes(x = qsec)) +
  geom_histogram(fill = "skyblue", color = "blue", bins = 10) +
  theme(axis.text = element_text(size = 16),
        axis.title = element_text(size = 16))
```

```r
ggplot(df, aes(x = qsec)) + 
    geom_density(color = "red") +
    labs(title = "Плотность распределения",
               x = "qsec",
               y = "Плотность") +
    theme(axis.text = element_text(size = 16),
            axis.title = element_text(size = 16))
```

2. Точечный график, на котором будет отображаться зависимость mpg от disp (plot). Задать свой цвет, форму элементов для каждого значения cyl. Размер элементов должен быть равен 2. 

```r
ggplot(mtcars, aes(x = disp, y = mpg, color = factor(cyl), shape = factor(cyl))) +
    geom_point(size = 2) +
    labs(title = "Зависимость mpg от disp",
           x = "Disp",
           y = "Mpg") +
      theme(axis.text = element_text(size = 16),
            axis.title = element_text(size = 16))
```

```r
ggplot(df, aes(x = factor(cyl), y = mpg)) +
  geom_boxplot(fill = "lightblue") +
  geom_jitter(width = 0.2, height = 0, color = "red") +
  labs(title = "Боксплоты и джиттерплоты mpg для каждого значения cyl",
       x = "Cylinder",
       y = "Mpg") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1),
        axis.text = element_text(size = 16),
        axis.title = element_text(size = 16))
```
